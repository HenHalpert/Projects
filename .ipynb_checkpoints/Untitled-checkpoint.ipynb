### Objective:
Analyzing the chemical abundances of Arcturus to understand the principles of optical spectroscopy and observations of stellar spectra. These principles will be applied by detecting the strengths of observed spectral lines, i.e the abundances of certain elements and comparing them to solar atmospheric model.

In Astronomy, astronomical instruments such as Optical spectrometers are used to analyze the radiation from astronomical objects and deduce chemical composition. Most of our knowledge regarding the chemical makeup of the universe comes from spectra. These measurements are useful for understanding the chemical evolution of the universe.
In addition to analyzing chemical compositions of astronomical objects, Spectroscopy can be used to derive many properties of distant stars and galaxies, such as temperature, density, mass, distance, luminosity, and relative motion using Doppler shift measurements. 
### Introduction:
The relative amount of elements in the atmosphere of a star can be determined by analyzing its stellar spectra, and measuring the strengths of absorption lines. The chemical abundances of a star are usually compared to a well-analyzed model, such as the atmospheric model of the Sun. Comparing the abundances of a star to the solar atmospheric model helps astronomers determine the age of the star, and its chemical evolution. The abundance of elements varies with the age and environment of the star. In order to understand nuclear fusion reactions that occur in a star, alpha processes elements are compared to the abundance of  iron, which are compared to the Sun([α/Fe]), and the iron abundance relative to Hydrogen is compared to the sun([Fe/H]). Figure (1) shown below is a plot of [α/Fe] vs [Fe/H]  of  arbitrary stars in our galaxy:
Figure (1): plot of [α/Fe] vs [Fe/H]  of  arbitrary stars in our galaxy(reference: lab manual)


The filled cirlces are a high-mass stars, and the triangles are low mass stars. The shapes in this graph are separated. That indicates that the mass of star tells us about the chemical abundances. Higher mass stars seem to have high constant [[α/Fe]] abundance from a some point around [Fe/H] = 1.0. Lower mass stars are mainly at the bottom right of the graph, which means that they have higher metallicity, and low alpha elements. These low mass stars produce mainly iron due to their evolution. Low mass stars can evolve into white dwarfs , and if it is in a binary system, a type Ia supernovae will occur, and iron will be produced, with almost no alpha elements production during this process. 
When High mass stars explode, they form Iron in the core, but they extract alpha elements into the interstellar medium. Hence, the production of new stars, will be comprised mostly from alpha elements. 
The strengths of spectral lines is usually determined by the equivalent width(EW)of the spectral line. It is measured by picking a continuum level, i.e a point at which the spectrum is continuum, that determines the level in which the area under the line will be taken for each absorption line. The EW's are usually approximated using normal distributions. However, the strength of the lines is not determined just by the abundance of each element. Parameters such as surface gravity, Temperature, and microturbulence, which are called the atmospheric parameters, effect the spectral lines as well.
